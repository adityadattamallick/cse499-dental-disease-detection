**Training with YOLOv8 Small Segmentation Model**

```python
import os
os.path.exists("/content/drive/MyDrive/research/base-172/data.yaml")
```

```python
from ultralytics import YOLO

# Load YOLOv8n-seg model
model = YOLO("/content/drive/MyDrive/yolov8s-seg.pt")  # pretrained on COCO

# Train the model
results = model.train(
    data="/content/drive/MyDrive/research/base-172/data.yaml",
    epochs=35,
    imgsz=640,
    batch=8,
    project="runs/train/base-172",
    name="base-172-run",
    exist_ok=True,
    task="segment",
    plots=True,
    auto_augment=False,
    augment=False,
    optimizer="SGD",       # SGD optimizer
    lr0=0.01,              # Initial learning rate
    lrf=0.001,             # Final learning rate
    patience=7,            # Patience for early stopping
    momentum=0.937,        # Momentum
    seed=42,
    save=True,
    val=True
)
```

**WandB Logging**

```python
import wandb
import os

RUN_DIR = "/content/runs/train/base-172/base-172-run"

wandb.init(project="research-base-172", entity="adityadatta30-north-south-university", name="research-base-172-results")

# Upload weights
wandb.save(os.path.join(RUN_DIR, "weights/best.pt"))
wandb.save(os.path.join(RUN_DIR, "weights/last.pt"))

# Upload CSV metrics
wandb.save(os.path.join(RUN_DIR, "results.csv"))

# Upload all PNG/JPG images
for file in os.listdir(RUN_DIR):
    if file.endswith((".png", ".jpg", ".jpeg")):
        wandb.log({file: wandb.Image(os.path.join(RUN_DIR, file))})

wandb.finish()
```
